```python
# ... Define functions for fraud detection
def find_fraud(data, features_to_check):
    pivot_stats=pd.pivot_table(data, values='transaction_id', aggfunc='count', index=features_to_check, columns='is_fraud', fill_value=0)
    pivot_stats.sort_values([True, False], ascending=False, inplace=True)
    pivot_stats.reset_index(inplace=True)
    pivot_stats['temp_fraud']=pivot_stats[True].map(lambda x: True if x>0 else False)
    pivot_stats.drop(columns=[True, False], inplace=True)
    data=trans.merge(pivot_stats, how='left', on=features_to_check)
    data['is_fraud']=data.temp_fraud
    data.drop(columns=['temp_fraud'], inplace=True)
    return data

def find_cc_fraud(data):
    data.sort_values(['cc_number', 'transaction_date'], inplace=True)
    first_fraud_date=data[['cc_number', 'transaction_date']].ix[data['is_fraud']==True].drop_duplicates('cc_number')
    first_fraud_date.columns=['cc_number', 'first_fraud_date']
    data=data.merge(first_fraud_date, how='left', on='cc_number')
    data['first_fraud_date']=data['first_fraud_date'].fillna(dt.datetime(2099,1,1))
    data['is_fraud'].ix[(data['transaction_date']>data['first_fraud_date']) &(data['is_fraud']==False)]=True
    data.drop(columns=['first_fraud_date'], inplace=True)
    return data
```

```python
# ... Run fraud detection algorithm
prev_num_fraud=0
num_fraud=(trans.is_fraud.map(lambda x: 1 if x else 0)).sum()

while num_fraud>prev_num_fraud:
    prev_num_fraud=num_fraud

    # ... Get stats per device id
    cols_to_check=['device_id']
    trans=find_fraud(trans, cols_to_check)    
    
    # ... Get stats per card
    trans=find_cc_fraud(trans) 
    
    # ... Get stats per card
    cols_to_check=['ip_address', 'cc_number']
    trans=find_fraud(trans, cols_to_check)
    
    # ... Get stats per card
    cols_to_check=['user_agent', 'cc_number']
    trans=find_fraud(trans, cols_to_check)
    
    # ... Get stats per card
    cols_to_check=['merchant', 'cc_number']
    trans=find_fraud(trans, cols_to_check)
    
    # ... Get stats per card
    cols_to_check=['email', 'cc_number']
    trans=find_fraud(trans, cols_to_check)
    
    # ... IP address and user agent
    cols_to_check=['ip_address', 'user_agent']
    trans=find_fraud(trans, cols_to_check)
    
    # ... IP address, shipping postal code and shipping phone number
    cols_to_check=['ip_address', 'shipping_phone_number']
    trans=find_fraud(trans, cols_to_check)
    
    # ... IP address, shipping postal code and shipping phone number
    cols_to_check=['ip_address', 'billing_phone_number']
    trans=find_fraud(trans, cols_to_check)
    
    # ... IP address and email
    cols_to_check=['ip_address', 'email']
    trans=find_fraud(trans, cols_to_check)
    
    # ... IP address and name
    cols_to_check=['email', 'user_agent']
    trans=find_fraud(trans, cols_to_check)
    
    num_fraud=(trans.is_fraud.map(lambda x: 1 if x else 0)).sum()
    print('iter')
```
